```python
# Someone on the webinar asked about using dates on an axis in Seaborn.  Here's an example.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
fire = pd.read_csv('london_fire.csv')
```

```python
# Checkt the types
fire.dtypes
```

```python
# Dates are not dates!  So let's convert them
fire.Date = pd.to_datetime(fire.Date)
```

```python
# Now plot
g = sns.relplot(x="Date", y="Grand Total", data=fire)
g.fig.set_size_inches(10,6)
plt.show()
```

```python
# Now plot
g = sns.relplot(kind="line", x="Date", y="Grand Total", data=fire)
g.fig.set_size_inches(10,6)
plt.show()
```

```python
# This plot messed up the date format, so I forced a format on it
g = sns.catplot(kind="bar", x="Date", y="Grand Total", data=fire, color="lightseagreen")
g.ax.set_xticklabels(labels=fire.Date.dt.strftime('%Y-%m-%d'), rotation=45)
g.ax.set_xticks(np.arange(0, fire.shape[0], 10))
g.fig.set_size_inches(18,6)
plt.show()
```
