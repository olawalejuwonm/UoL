# Seaborn 3: Univariate Distributions

Understanding the distribution of a single feature is an important part of data analysis.  Seaborn provides the [distplot](https://seaborn.pydata.org/generated/seaborn.distplot.html) function for this purpose.

```python
# Load numpy and pandas for list and data set manipulation
import numpy as np
import pandas as pd

# Load matplotlib and seaborn for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style for Seaborn
sns.set(color_codes=True)
```

```python
world_data = pd.read_csv('world_data.csv')
```

We will focus on plotting the inequality indicator. We need to remove nulls from this data or some of the plots will fail.

```python
inequality = world_data.inequality.dropna()
```

## Univariate distribution

Let's plot the distribution of inequality. 

```python
sns.distplot(inequality)
```

We can see a histogram and a [kernel density estimate](https://mathisonian.github.io/kde/) (KDE) line.  The KDE provides an estimate of the probability distribution of the data.  Effectively, it is smoothing the histogram.

We can remove the kernel density estimate line:

```python
sns.distplot(inequality, kde=False)
```

Or we can remove the histogram:

```python
sns.distplot(inequality, hist=False)
```

We can add a tick per observation (this is called a rugplot).

```python
sns.distplot(inequality, kde=False, rug=True)
```

We can change the number of bins.

```python
sns.distplot(inequality, bins=20, kde=False)
```

If you just want a KDE plot, use kdeplot as it gives some more options:

```python
sns.kdeplot(inequality, shade=True)
```

For example, we can change the bandwidth to affect the leve of smoothing:

```python
sns.kdeplot(inequality, bw=0.1)
```

## Fitting parametric distributions
We can also fit a standard parametric distribution to the data to see how closely it matches.  This gives us an insight into how we might make generalisations about our data, which can be used to make statistical inferences.

For example, here we can see how well our data fits a normal distribution:

```python
from scipy import stats
sns.distplot(inequality, kde=False, fit=stats.gennorm)
```

Or even something more unusual, like a triangular distribution:

```python
sns.distplot(inequality, kde=False, fit=stats.triang)
```

There are a whole wealth of distributions:
https://docs.scipy.org/doc/scipy/reference/stats.html
