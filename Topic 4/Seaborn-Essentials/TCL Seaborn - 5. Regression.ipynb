# Seaborn 5: Regression

When analysing numeric data, it can be helpful to estimate the relationship between two features.  Seaborn provides the ability to quickly estimate and visualise these relationhips.  Note, however, this should not be used as a substitute for building a full statisical model.

We can use the Seaborn [regplot](https://seaborn.pydata.org/generated/seaborn.regplot.html) function.

```python
# Load numpy and pandas for list and data set manipulation
import numpy as np
import pandas as pd

# Load matplotlib and seaborn for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style for Seaborn
sns.set(color_codes=True)
```

```python
world_data = pd.read_csv('world_data.csv')
```

## Linear regression
Let's say we want to examine the relationship between sanitation and life expectancy, perhaps with a view to using sanitation to predict life expectancy.

```python
sns.regplot(x="sanitation", y="lifeexp", data=world_data)
```

The shaded area provides a 95% confidence interval.

The above gives a simple linear regression. It fits a straight line to the data.  Sometimes a straight line is not enough!  We can apply **polynomial regression** by specifying the order.  Unlike linear regression, polynomial regression lines can bend!  The higher the order, the more the line bends to fit the data.  Don't go too mad.  The ideal line is the simplest one that models the general trend in the data, not the one that captures the most points!

```python
sns.lmplot(x="sanitation", y="lifeexp", order=2, data=world_data)
```

```python
sns.lmplot(x="foodsupply", y="childmortality", data=world_data)
```

```python
sns.lmplot(x="foodsupply", y="childmortality", order=2, data=world_data)
```

## Logistic regression
Where we have a binary feature on y, we can try to fit a linear regression line.  Here we have a binary flag for high life expectancy on the y axis:

```python
sns.lmplot(x="foodsupply", y="highlifeexp", data=world_data)
```

As you can see, it's drawn a line, but it doesn't really fit.  

A logistic regression will provide a better fit:

```python
sns.lmplot(x="foodsupply", y="highlifeexp", logistic=True, data=world_data)
```

We can think of the above curve (called a sigmoid curve) as the probability of a country having a high life expectancy based on the food supply.  The higher the food supply, the more likely the country will be classified as having a high life expectancy.

## Cutting by another feature
We can use another feature as the hue parameter to split the data and see the regression lines for the different classes for that feature.  Eg here we can see that higher inequality increases childmortaility at a higher rate for low income countries.

```python
sns.lmplot(x="inequality", y="childmortality", hue='highincome', data=world_data)
```

And let's split further by high and low life expectancy:

```python
sns.lmplot(x="inequality", y="childmortality", hue='highincome', col="highlifeexp", data=world_data)
```
