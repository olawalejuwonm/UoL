# Matplotlib: Additional Techniques
In this notebook I'll show a few additional techniques that can be used with Matplotlib.

```python
import matplotlib.pyplot as plt
```

## Combining charts
If you wish, you can combine different types of chart

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [1,2,3,4,5]
y1 = [0,20,30,25,15]                                      # line data
y2 = [0,5, 12, 27, 35]                                    # bar data

# Plot the data
plt.plot(x, y1, label='Mild Cheddar')                     # plot line
plt.bar(x, y2, label='Blue Thunder', color='lightcoral')  # plot bar

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")
plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```

## Annotations
You can add annotations to your chart using the [**annotate()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html) method.  Provide the text and position of the annotated point  as a minimum.  The coordinates are, by default, in data units.  You can offset the text, so it doesn't clash with other text or graphics, and add arrows.

```python
#help(plt.annotate)
```

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]

# Plot the data
plt.plot(x, y, label='Mild Cheddar')                      # add a label


# Add title and axis labels
plt.title("Cheese flavour over time")                     # add a title
plt.xlabel("Cheese age (years)")                          # add an x-axis label
plt.ylabel("Tastiness score")                             # add a y-axis label

# Label a data point
plt.annotate("tasty", xy=(x[1],y[1]))

# Can shift the position relative to the data point using xytext
plt.annotate("very, very tasty", xy=(x[2],y[2]), xytext=(x[2]-1.25,y[2]-1), color='g')

# Can add an arrow
plt.annotate("very tasty", xy=(x[3],y[3]), xytext=(x[3]+0.25,y[3]+2), arrowprops={'width':1}, color='g')

# Show the plot
plt.show()
```

## Horizontal and vertical lines
Sometimes you want to add a horizontal or vertical line to identify a particular x or y value.  Use the [**axhline()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axhline.html) or [**axvline()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axvline.html) methods.

```python
#help(plt.axhline)
#help(plt.axvline)
```

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]

# Plot the data
plt.plot(x, y, label='Mild Cheddar')                      # add a label


# Add title and axis labels
plt.title("Cheese flavour over time")                     # add a title
plt.xlabel("Cheese age (years)")                          # add an x-axis label
plt.ylabel("Tastiness score")                             # add a y-axis label

# Horiztonal line
plt.axhline(y=25, color='r', linestyle='--')

# Vertical line
plt.axvline(x=3.5, color='g', linestyle=':')

# Show the plot
plt.show()
```

## Shading areas
You can shade area of the chart using the [**fill_between()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html) method.  

```python
#help(plt.fill_between)
```

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]

# Plot the data
plt.plot(x, y, label='Mild Cheddar')                      # add a label


# Add title and axis labels
plt.title("Cheese flavour over time")                     # add a title
plt.xlabel("Cheese age (years)")                          # add an x-axis label
plt.ylabel("Tastiness score")                             # add a y-axis label

# Select first 3 data points
plt.fill_between(x[:4], 0, y[:4], facecolor='pink')

# Show the plot
plt.show()
```

We can use a condition to decide which areas to fill.  E.g. here we select datapoints at either end of the chart

```python
# Import the library
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data
x = np.array([0, 1, 2, 3, 4])
y = [0,20,30,25,15]

# Plot the data
plt.plot(x, y, label='Mild Cheddar')                      # add a label


# Add title and axis labels
plt.title("Cheese flavour over time")                     # add a title
plt.xlabel("Cheese age (years)")                          # add an x-axis label
plt.ylabel("Tastiness score")                             # add a y-axis label

# Select all data points but use a condition
condition = (x<=1) | (x>=3) # this generates [True,True,False,True,True]
plt.fill_between(x, 0, y, where=condition, facecolor='coral')


# Show the plot
plt.show()
```

This is the example from the Matplotlib documentation

```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(0,10,0.1)
  
# plotting the lines
a1 = 4 - 2*x
a2 = 3 - 0.5*x
a3 = 1 -x
  
# The upper edge of polygon
a4 = np.minimum(a1, a2)
  
# Setting the y-limit
plt.ylim(0, 5)
  
# Plot the lines
plt.plot(x, a1,
        x, a2,
        x, a3)
  
# Filling between line a3 and line a4
plt.fill_between(x, a3, a4, color='green',
                 alpha=0.5)
plt.show()
```
