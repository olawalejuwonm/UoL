# Plotting with Pandas
Pandas has a convenience wrapper on matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fire = pd.read_csv('london_fire_brigade.csv')
```

```python
fire.head()
```

## Simple plot
### Matplotlib way
We just pass the series as the x and y

```python
plt.plot(fire.Date, fire.Incidents)
plt.show()
```

```python
plt.bar(fire.Date, fire.Incidents)
plt.show()
```

### Pandas way
We make the method call on the dataframe and pass the column names.

```python
fire.plot(x="Date", y="Incidents")
plt.show()
```
