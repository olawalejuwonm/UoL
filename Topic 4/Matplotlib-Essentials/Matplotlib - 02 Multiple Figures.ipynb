We can create a grid of mulitple axes.  This grid is called a set of **subplots**.

Let's create a **figure** that is a 2x3 grid of axes:




Remember that **axes** is the term used to refer to an individual chart.

There are a couple of ways to create these sets of subplots.

## Method 1: Use the subplot() function

First we create the [**figure()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) that will contain the axes.  The size is the size in inches:

```python
fig = plt.figure(figsize=(12, 8))
```

Each axes has an index position, as indicated by the numbers in the picture above.

The [**subplot()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) function selects an axes, setting it as the current axes.  

```python
plt.subplot(231)
```

All plotting will be in that axes until we change axes.

The 3-digit number in the call to subplot() is actually 3 separate numbers representing the number of **rows**, number of **columns** and **index**.  

E.g. subplot(231) means we have 2 rows, 3 columns in our plot grid and we are currently drawing axes 1.

````python
import matplotlib.pyplot as plt

# Prepare the data
x = [1, 2, 3, 4, 5]
y1 = [0, 20, 30, 25,15]                       
y2 = [0, 5, 12, 27, 35]                               
y3 = [0, 2, 7, 12, 35]
y4 = [0, 4, 6, 12, 12]
y5 = [0, 5, 6, 8, 10]
y6 = [0, 6, 12, 15, 19]

# Create the figure that will contain the axes
fig = plt.figure(figsize=(12, 8))

plt.subplot(231)
plt.plot(x, y1, label='Mild Cheddar')  
plt.legend()

plt.subplot(232)
plt.plot(x, y2, label='Blue Thunder')
plt.legend()

plt.subplot(233)
plt.plot(x, y3, label='Gorgonzola') 
plt.legend()

plt.subplot(234)
plt.plot(x, y4, label='Brie')
plt.legend()

plt.subplot(235)
plt.plot(x, y5, label='Comte') 
plt.legend()

plt.subplot(236)
plt.plot(x, y6, label='Manchego')
plt.legend()

# Show the plot
plt.show()
````

It's useful to confirm that the figure is made up of a number of axes.

````python
fig.get_axes()
````

## Method 2: Use the subplots() function
This method achieves the same goal but uses a different syntax.  First we specify the grid dimensions using the [**subplots()**](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function:

```python
fig, ax = plt.subplots(2,3)
```

The return value is a figure and a list of axes.

Then, when we want to draw a figure, we select the axes for the figure and make our plot:


```python
ax[0,0].plot(x, y1, label='Mild Cheddar') 
ax[0,0].legend()
```



````python
import matplotlib.pyplot as plt

# Prepare the data
x = [1, 2, 3, 4, 5]
y1 = [0, 20, 30, 25,15]                       
y2 = [0, 5, 12, 27, 35]                               
y3 = [0, 2, 7, 12, 35]
y4 = [0, 4, 6, 12, 12]
y5 = [0, 5, 6, 8, 10]
y6 = [0, 6, 12, 15, 19]

# Create a grid with 2 rows and 3 columns
fig, ax = plt.subplots(2,3, figsize=(12, 8))

ax[0,0].plot(x, y1, label='Mild Cheddar') 
ax[0,0].legend()

ax[0,1].plot(x, y2, label='Blue Thunder')
ax[0,1].legend()

ax[0,2].plot(x, y3, label='Gorgonzola') 
ax[0,2].legend()

ax[1,0].plot(x, y4, label='Brie')
ax[1,0].legend()

ax[1,1].plot(x, y5, label='Comte') 
ax[1,1].legend()

ax[1,2].plot(x, y6, label='Manchego')
ax[1,2].legend()

plt.show()
````

## Spanning rows and columns
Sometimes we want figures to span across a number of rows or columns in our grid.  We can use the [**subplot2grid()**](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplot2grid.html) function for this.

````python
# Number of rows and columns
gridsize = (3, 2)

# Create the figure that will contain the axes
fig = plt.figure(figsize=(12, 8))

# Create an axes that spans 2 rows and 2 columns
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=2, rowspan=2)

# Create two axes that occupy the other 2 positions in the grid
ax2 = plt.subplot2grid(gridsize, (2, 0))
ax3 = plt.subplot2grid(gridsize, (2, 1))

# Now we can plot each axes
# ...
 
````
