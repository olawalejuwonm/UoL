# Visualisation with Matplotlib

Matplotlib is the core data visualisation library for Python.  A lot of other libraries (such as Seaborn) are built on top of it.  It is worth getting familiar with matplotlib as it crops up a lot!

This is the standard way to import matplotlib:

```python
# Import the library
import matplotlib.pyplot as plt
import numpy as np
```

# A quick and simple plot

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]

# Plot the data (by default it plots a line chart)
plt.plot(x, y)

# Show the plot
plt.show()
```

## Add annotations
Let's add some simple annotations to the plot:

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]

# Plot the data
plt.plot(x, y, label='Mild Cheddar')                      # add a label

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")                     # add a title
plt.xlabel("Cheese age (years)")                          # add an x-axis label
plt.ylabel("Tastiness score")                             # add a y-axis label

# Show the plot
plt.show()
```

## Add another line
We can add a second line to the same plot:

```python
# Import the library
import matplotlib.pyplot as plt

# Prepare the data
x = [1,2,3,4,5]
y1 = [0,20,30,25,15]                                      # line 1 data
y2 = [0,5, 12, 27, 35]                                    # line 2 data

# Plot the data
plt.plot(x, y1, label='Mild Cheddar')                     # plot line 1
plt.plot(x, y2, label='Blue Thunder')                     # plot line 2

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")
plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```

## Add a style
We can use some built-in styles

```python
# Import the library
import matplotlib.pyplot as plt
from matplotlib import style                              # import the Matplotlib style library

style.use('ggplot')                                       # select the ggplot style

# Prepare the data
x = [1,2,3,4,5]
y1 = [0,20,30,25,15]
y2 = [0,5, 12, 27, 35]

# Plot the data
plt.plot(x, y1, label='Mild Cheddar')
plt.plot(x, y2, label='Blue Thunder')

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")
plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```

## Colours
We can give each line a defined colour, overriding the built-in colours:

```python
# Import the library
import matplotlib.pyplot as plt
from matplotlib import style                              # import the Matplotlib style library

style.use('ggplot')                                       # select the ggplot style

# Prepare the data
x = [1,2,3,4,5]
y1 = [0,20,30,25,15]
y2 = [0,5, 12, 27, 35]

# Plot the data
plt.plot(x, y1, label='Mild Cheddar', color='teal')           # specify colour for line
plt.plot(x, y2, label='Blue Thunder', color='lightcoral')     # specify colour for line

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")

plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```

Take a look at the possible named colours and experiment with them:

https://matplotlib.org/gallery/color/named_colors.html


## Scatter plot
Instead of using plt.plot() for line plots, we can use **plt.scatter()** for scatter plots of one feature against another:

```python
# Import the library
import matplotlib.pyplot as plt
from matplotlib import style  

style.use('ggplot') 

# Prepare the data
x = [1,2,3,4,5]
y1 = [0,20,30,25,15]
y2 = [0,5, 12, 27, 35]

# Plot the data
plt.scatter(x, y1, label='Mild Cheddar', color='teal')         # change to scatter plot
plt.scatter(x, y2, label='Blue Thunder', color='lightcoral')   # change to scatter plot 

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")

plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```

## Bar charts
We can use **plt.bar()** to create bar charts.

```python
# Import the library
import matplotlib.pyplot as plt
from matplotlib import style  

style.use('ggplot') 

# Prepare the data
x = [1,2,3,4,5]
y1 = [0,20,30,25,15]
y2 = [0,5, 12, 27, 35]

# Plot the data
plt.bar(x, y1, label='Mild Cheddar', color='teal')         # change to a bar
plt.bar(x, y2, label='Blue Thunder', color='lightcoral')   # change to a bar

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")

plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```

We see a problem here.  The bars are overlapping, so the Blue Thunder bars are drawn over the Mild Cheddar bars.  Let's fix that by resizing and moving the bar:

```python
x = np.array([1,2,3,4,5])  
x - 0.4
```

```python
# Import the library
import matplotlib.pyplot as plt
from matplotlib import style  
import numpy as np

style.use('ggplot') 

# Prepare the data
x = np.array([1,2,3,4,5])                                                               # use numpy so we can perform arithmetic on the whole list at once
y1 = [0,20,30,25,15]
y2 = [0,5, 12, 27, 35]

bar_width = 0.4                                                                         # set width to be 0.4 x width of total original bar

# Plot the data
plt.bar(x-bar_width/2, y1, label='Mild Cheddar', width=bar_width, color='teal')         # change width and move backwards
plt.bar(x+bar_width/2, y2, label='Blue Thunder', width=bar_width, color='lightcoral')   # change width and move forwards

# Add a legend
plt.legend()

# Add title and axis labels
plt.title("Cheese flavour over time")

plt.xlabel("Cheese age (years)")
plt.ylabel("Tastiness score")

# Show the plot
plt.show()
```
