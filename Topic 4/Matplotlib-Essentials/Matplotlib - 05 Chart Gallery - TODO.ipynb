```python
# Import the library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Simple line

```python
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]
plt.plot(x, y)
plt.show()
```

## Simple bar

```python
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]
plt.bar(x, y)
plt.show()
```

## Simple horizontal bar

```python
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]
plt.barh(x, y)
plt.show()
```

## Horizontal bar with error bars

```python
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]
err = [4,7,3,6,9]
plt.barh(x, y,xerr=err)
plt.show()
```

## Stacked bars

```python
x = [0, 1, 2, 3, 4]
y1 = [0,20,30,25,15] 
y2 = [0,5, 12, 27, 35]  

plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.show()
```

## Labelled bars

```python
x = [0, 1, 2, 3, 4]
y = [0,20,30,25,15]

p1 = plt.bar(x, y)
plt.bar_label(p1)
plt.show()
```

## Sized bubbles

```python
x = [0, 1, 2, 3, 4]
y1 = [0,20,30,25,15] 
y2 = [200, 400, 12, 160, 35] 
plt.scatter(x, y1, s=y2)
plt.show()
```

## Coloured bubbles

```python
x = [0, 1, 2, 3, 4]
y1 = [0,20,30,25,15] 
y2 = [0, 5, 12, 30, 35] 
plt.scatter(x, y1, c=y2)
plt.show()
```

## Stackplot

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = {
    'a': [228, 284, 365, 477, 631, 814, 1044, 1275],
    'b': [340, 425, 519, 619, 727, 840, 943, 1006],
    'c': [1394, 1686, 2120, 2625, 3202, 3714, 4169, 4560],
    'd': [220, 253, 276, 295, 310, 303, 294, 293],
    'e': [12, 15, 19, 22, 26, 31, 36, 39],
}

plt.stackplot(x, y.values())
plt.show()
```
