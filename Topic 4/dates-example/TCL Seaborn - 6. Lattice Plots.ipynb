# Seaborn 6: Building up lattice plots 

Lattice plots are so useful that it is worth understanding how to build them up from scratch, so you can get the ultimate control.

This approach is based on Seaborn's [FacetGrid](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html) function.

```python
# Load numpy and pandas for list and data set manipulation
import numpy as np
import pandas as pd

# Load matplotlib and seaborn for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
world_data = pd.read_csv('world_data.csv')
```

First, we can create a facet grid for our data.  This sets up the grid without plotting any data:

```python
g = sns.FacetGrid(world_data, col="happinessband")
```

We can now "map" a feature to a plotting function, applying the plotting function to each facet's subset of the data:

```python
g = sns.FacetGrid(world_data, col="happinessband")
g.map(plt.hist, "income")
```

We can build up plots in this way.  The first parameter of [map](
https://seaborn.pydata.org/generated/seaborn.FacetGrid.map.html) is the function to call.  Subsequent parameters are columns to pass to the function, in the order specified by the function.

For example, we can use the scatter function as the first parameter.  If you look up the [documentation for scatter](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html), you will see that we need to provide at least the x and y features. So we will pass the x and y features as the subsequent parameters:




```python
g = sns.FacetGrid(world_data, col="happinessband", hue="region")
g.map(plt.scatter, "unemployment", "foodsupply", alpha=.7)
g.add_legend();
```

Let's try another one:

```python
g = sns.FacetGrid(world_data, col="lifeexpband", row="region", col_order=['L','M','H'], aspect=2)
g.map(sns.pointplot, "happinessband", "unemployment", order=[1, 2, 3, 4], alpha=.7)
g.add_legend();
```

For more information and ideas, see the Seaborn tutorial:

https://seaborn.pydata.org/tutorial/axis_grids.html
