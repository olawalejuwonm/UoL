# Seaborn 7: Styling
Making nice-looking plots is important to engage your audience.  But good selection of stylistic chart elements can go a long way towards helping your audience *understand* your plot.  Let's look at some ways to customise your chart styles.

```python
# Load numpy and pandas for list and data set manipulation
import numpy as np
import pandas as pd

# Load matplotlib and seaborn for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
world_data = pd.read_csv('world_data.csv')
```

## Style
Seaborn has 5 pre-set styles that can be used to quickly change the general look.  We use the [set_style](https://seaborn.pydata.org/generated/seaborn.set_style.html) function.

### Whitegrid

```python
sns.set_style("whitegrid")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Blackgrid

```python
sns.set_style("darkgrid")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### White

```python
sns.set_style("white")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Dark

```python
sns.set_style("dark")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Ticks
Adds small tick marks on the axes.

```python
sns.set_style("ticks")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Overriding individual style elements
You can override individual styles in the call to set_style, by providing the style name and setting in curly brackets:

```python
sns.set_style("whitegrid", {"axes.facecolor": "#eeeeff"})
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

You can run the following command to see a list of styles that can be overridden in this way:

```python
sns.axes_style()
```

### Context
You can scale the lines, fonts and other plot elements, whilst keeping the size of the plot the same. There are 4 sizes:


paper < notebook < talk < poster


```python
# Reset styles
sns.set()
```

### Paper

```python
sns.set_context("paper")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Notebook

```python
sns.set_context("notebook")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Talk

```python
sns.set_context("talk")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Poster

```python
sns.set_context("poster")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Overriding individual context elements

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 5})
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

## Colour palettes
Seaborn has a huge scope for customising the colours of your plot.

```python
# Reset styles
sns.set()

sns.set_style("whitegrid")
```

### Categorical vs Sequential vs Divergent
It is important to choose a colour scheme that aids understanding of your visualisations.  There are 3 classes of colour palette that we use:

Categorical colour palettes are useful for categorical data.  Each colour is very distinctive and there is no relationship between one colour and another:

```python
p = sns.color_palette("bright",10)
sns.palplot(p)
```

Use categorical palettes to plot categorical / qualitative data:

```python
sns.set_palette("bright",10)
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

Sequential colour palettes are used for showing a transition from low to high values

```python
p = sns.color_palette("Reds", 10)
sns.palplot(p)
```

Use sequential palettes for things like KDE plots or heat maps:

```python
ile = world_data[world_data.region=="Europe & Central Asia"].dropna()
sns.kdeplot(ile.income, ile.lifeexp, shade=True, cmap="Reds", shade_lowest=False)
```

Diverging colour palettes can be used to highlight values at the ends of a scale (high or low):

```python
p = sns.color_palette("RdBu", 10)
sns.palplot(p)
```

Diverging colour palettes work really rell for heat maps.  Here, we have a correlation map, showing which features are strongly correlated in our data.  Dark blue signifies a strong positive correlation and dark red signifies a strong negative correlation.

```python
sns.heatmap(world_data.corr(), cmap="RdBu")
```

### Default categorical colours

We can set the colour palette to use for a plot using the [set_palette](https://seaborn.pydata.org/generated/seaborn.set_palette.html) function:

```python
sns.set_palette("pastel")
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

### Circular categorical colours - hls
Let's say you have a chart where you need 6 categorical colours.  You can generate a palette of evenly spaced colours:

```python
p = sns.color_palette("hls", 6)
sns.palplot(p)
```

You can generate as many as you want in this way.

```python
p = sns.color_palette("hls", 12)
sns.palplot(p)
```

Note how the colours are effively following the rainbow colours! ROYGBIV.

### Circular categorical colours - husl
THe [HSLuv](https://www.hsluv.org/) system improves the preceived intensity of the chosen colours, so is generally a better bet:

```python
p = sns.color_palette("husl", 6)
sns.palplot(p)
```

```python
p = sns.color_palette("husl", 12)
sns.palplot(p)
```

### Color Brewer colours

[ColourBrewer](http://colorbrewer2.org/) is a great colour selection tool.  Seaborn has ColourBrewer palettes built-in, e.g.:

```python
p = sns.color_palette("Paired", 12)
sns.palplot(p)
```

```python
p = sns.color_palette("Set2", 12)
sns.palplot(p)
```

Seaborn provides the [choose_colorbrewer_palette](https://seaborn.pydata.org/generated/seaborn.choose_colorbrewer_palette.html) function, which allows you to interactively define Color Brewer palettes

```python
sns.choose_colorbrewer_palette('qualitative')
```

```python
sns.choose_colorbrewer_palette('diverging')
```

```python
sns.choose_colorbrewer_palette('sequential')
```

Once you have designed your palette, you can create a palette object and set it:

```python
p = sns.color_palette("Accent", n_colors=8, desat=1)
sns.set_palette(p)
```

```python
sns.catplot(kind="bar", y="region", x="energy", orient="h", aspect=2, data=world_data)
```

## Complete customisation
You can also create your very own custom colour palette by specifying the individual colours.

```python
colors = ["#94A2F2", "#0D0D0D", "#6A7329", "#F2E85C", "#F25922"]
p = sns.color_palette(colors)
sns.palplot(p)
```
